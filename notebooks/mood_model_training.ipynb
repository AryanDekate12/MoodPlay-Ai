# Import Liabraries

```python
import pandas as pd
import pickle

from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
```

# Create Training Data

```python
data = {
    "mood":[
        "happy","happy",
        "sad","sad",
        "angry","angry",
        "bored","bored",
        "relaxed","relaxed"
    ],
    
    "genre":[
        "sports","adventure",
        "simulation","story-rich",
        "shooter","action",
        "puzzle","arcade",
        "casual","simulation"
    ]
}

df = pd.DataFrame(data)

df
```

# Encode Text

```python
mood_encoder = LabelEncoder()
genre_encoder = LabelEncoder()

df["mood_encoded"] = mood_encoder.fit_transform(df["mood"])
df["genre_encoded"] = genre_encoder.fit_transform(df["genre"])

df
```

# Prepare Training Data

```python
X = df[["mood_encoded"]]
y = df["genre_encoded"]
```

# Train Model

```python
model = DecisionTreeClassifier()

model.fit(X,y)

print("Model trained successfully")
```

# Save Model

```python
pickle.dump(model, open(r"C:\Users\aryan\practice projects\MoodPlay Ai\models\mood_model.pkl", "wb"))
pickle.dump(mood_encoder, open(r"C:\Users\aryan\practice projects\MoodPlay Ai\models\mood_encoder.pkl", "wb"))
pickle.dump(genre_encoder, open(r"C:\Users\aryan\practice projects\MoodPlay Ai\models\genre_encoder.pkl", "wb"))

print("Model saved successfully")
```
